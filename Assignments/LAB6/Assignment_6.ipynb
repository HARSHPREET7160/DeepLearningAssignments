```python
import numpy as np
```

```python
X = np.linspace(0, 10, 50).reshape(-1, 1)
```

```python
noise = np.random.normal(0, 0.2, size=(50, 1))
y = np.log(X + 1) + noise
```

```python
# choose 3 hidden units to give flexibility to bend the curve
# More than 1 is needed because one unit is too limited
# Too many units can overcomplicate training.
hidden_units = 3
```

```python
W1 = np.random.uniform(-1, 1, size=(1, hidden_units))
b1 = np.zeros((1, hidden_units))
W2 = np.random.uniform(-1, 1, size=(hidden_units, 1))
b2 = np.zeros((1, 1))
```

```python
def activation(z):
    return np.maximum(0, z)

def activation_slope(z):
    return (z > 0).astype(float)

# The slope tells learning where changes are possible, so gradients can flow backward.
```

```python
z1 = X @ W1 + b1
h = activation(z1)
y_hat = h @ W2 + b2
```

```python
error = y_hat - y
loss = np.mean(error ** 2)
# Squaring makes large errors contribute much more than small errors
```

```python
# To reduce loss, we must change parameters.
# To know how to change them, we need to know if a small change increases or decreases loss.
```

```python
dL_dy = 2 * error / len(X)
dL_dW2 = h.T @ dL_dy
dL_db2 = np.sum(dL_dy, axis=0, keepdims=True)
# These depend on hidden values because output weights connect hidden activations to prediction.

dL_dh = dL_dy @ W2.T
dL_dz1 = dL_dh * activation_slope(z1)
# Activation slope gates error flow: positive z passes gradient, non-positive z blocks it.

dL_dW1 = X.T @ dL_dz1
dL_db1 = np.sum(dL_dz1, axis=0, keepdims=True)
```

```python
learning_rate = 0.01
W1 -= learning_rate * dL_dW1
b1 -= learning_rate * dL_db1
W2 -= learning_rate * dL_dW2
b2 -= learning_rate * dL_db2
# This is like perceptron-style update: move parameters opposite to the error gradient to reduce loss.
```

```python
epochs = 1000
learning_rate = 0.01
# 1000 epochs so the model has enough passes to gradually reduce loss.
# learning rate 0.01 to make updates steady without jumping too much.

for epoch in range(epochs):
    z1 = X @ W1 + b1
    h = activation(z1)
    y_hat = h @ W2 + b2

    error = y_hat - y
    loss = np.mean(error ** 2)

    dL_dy = 2 * error / len(X)
    dL_dW2 = h.T @ dL_dy
    dL_db2 = np.sum(dL_dy, axis=0, keepdims=True)

    dL_dh = dL_dy @ W2.T
    dL_dz1 = dL_dh * activation_slope(z1)
    dL_dW1 = X.T @ dL_dz1
    dL_db1 = np.sum(dL_dz1, axis=0, keepdims=True)

    W1 -= learning_rate * dL_dW1
    b1 -= learning_rate * dL_db1
    W2 -= learning_rate * dL_dW2
    b2 -= learning_rate * dL_db2

    if epoch % 100 == 0:
        print(epoch, loss)
```
